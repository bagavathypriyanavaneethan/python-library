# Sorting dataframe or series

```python
import pandas as pd
```

```python
df = pd.read_csv('http://bit.ly/imdbratings')
```

```python
df.head()
```

```python
df.title.sort_values()
```

### Here first number wise sorted & alphabetical order

```python
df['genre'].sort_values()
```

```python
type(df['genre'].sort_values())
```

### So, the sort_values() return pandas series

## Descending order

```python
df['genre'].sort_values(ascending = False)
```

## Arrange the dataframe by single field

```python
df.sort_values('genre')
```

**Here all the values are arranged by the genre alphabetically**

```python
type(df.sort_values('genre'))
```

Now it returns the entire data frame

```python
df.sort_values('duration',ascending=False)
```

## Sort with two columns

```python
df.sort_values(['content_rating','genre'])
```

**Here first content rating is sorted & values are sorted within genre field**
