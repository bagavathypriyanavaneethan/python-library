# Selecting multiple rows & columns

```python
import pandas as pd
```

```python
ufo = pd.read_csv("http://bit.ly/uforeports")
```

```python
ufo.head()
```

### loc is used to select rows using the syntax [row, column]

```python
ufo.loc[0,:]
```

```python
ufo.loc[0:5,:]
```

```python
ufo.loc[[0,1,2,3],:]
```

```python
ufo.loc[:,'City':'State']
```

```python
ufo.head(3).drop('State',axis=1)
```

# Using boolean expression 

```python
ufo.loc[ufo.City=='Holyoke',:]
```

```python
ufo[ufo.City=='Ithaca'].State  #Chain indexing but it may cause problem
```

### iloc - integer location 

#### it uses index to select the data

```python
ufo.iloc[0,4]
```

```python
ufo.iloc[0:5,:]
```

#### The iloc function excludes the last no in the range specified (here 5 as 0:5), but the loc function includes from the start to end in the index

```python
ufo[['City','State']]  #implicit : Not recommended

ufo.loc[:,['City','State']]  #explicitly defining the row & column values
```

```python
ufo[0:2]   #Not recommended

ufo.iloc[0:2,:]
```
